```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
data = pd.read_csv('advertising.csv')
data.head()
```

```python
data.drop(['TV', 'Radio'], axis=1, inplace=True)
data.head()
```

```python
dend = shc.dendrogram(shc.linkage(data, method='ward', metric='euclidean'))
```

```python
agnes = AgglomerativeClustering(n_clusters=5, linkage='ward', metric='euclidean')
agnes.fit(data)
```

```python
data_labels = agnes.labels_
sns.scatterplot(x='Newspaper', y='Sales', data=data, hue=data_labels).set_title('Agnes')
```

```python
dend1 = shc.dendrogram(shc.linkage(data, method='complete', metric='euclidean'))
```

```python
diana = AgglomerativeClustering(n_clusters=5, linkage='complete', metric='euclidean')
diana.fit(data)
```

```python
data_labels2 = diana.labels_
sns.scatterplot(x='Newspaper', y='Sales', data=data, hue=data_labels2).set_title("DIANA")
```
