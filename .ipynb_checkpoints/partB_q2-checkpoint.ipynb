```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('advertising.csv')
data.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop(['Sales'], axis=1), data['Sales'], test_size=0.2)

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print("Coeff = {} \nIntercept = {}".format(lr.coef_, lr.intercept_))
print("Line : Y= {:.3}*TV + {:.2}*Radio + {:.3}*Newspaper + {:.3}".format(lr.coef_[0], lr.coef_[1], lr.coef_[2], lr.intercept_))
```

```python
y_pred = lr.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
score = lr.score(data.drop(['Sales'], axis=1), data['Sales'])

print(f"rmse = {rmse} \nmae = {mae} \nScore = {score}")
```

```python
def trainOnly2Features(f1,f2):
    
    print(("Only taking features "+f1+" & "+f2).center(50,'='))
    x_train,x_test,y_train,y_test=train_test_split(data[[f1,f2]],data['Sales'],test_size=0.25)
    lr = LinearRegression()
    lr.fit(x_train,y_train)
    
    print("Coeffs of the equation = {} \nIntercept = {} ".format(lr.coef_,lr.intercept_))
    print("The linear model is: Y = {:.5} + {:.5}*{} + {:.5}*{} ".format(lr.intercept_, lr.coef_[0],f1, lr.coef_[1],f2))
    pred = lr.predict(x_test)
    
    
    rmse=np.sqrt(mean_squared_error(y_test,pred))
    mae = mean_absolute_error(y_test,pred)
    mse = mean_squared_error(y_test,pred)
    
    print("Root Mean Square Srror= {}\nMean Absolute Error = {}\nMean Squared Error = {}".format(rmse,mae,mse))
```

```python
trainOnly2Features('TV', 'Radio')
```
